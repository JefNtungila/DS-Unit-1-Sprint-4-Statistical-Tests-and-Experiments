##TTEST

```
from scipy.stats import ttest_ind, ttest_ind_from_stats, ttest_rel
import numpy as np
```

```
positive_credit = credit_data[credit_data['A16']==1]
positive_credit.head()
```

```
positive_credit_A8 = positive_credit['A8']

```

```
negative_credit = credit_data[credit_data['A16']!=1]
negative_credit.head()
```

```
negative_credit_A8 = negative_credit['A8']
```

```
for deciding_factor in [positive_credit_A8, negative_credit_A8]:
    print(f"Mean: {deciding_factor.mean()}")
    print(f"Standard Deviation: {deciding_factor.std()}")
    print(f"Variance: {deciding_factor.var()}")
    print("---"*10)
```

```
statistic, pvalue = ttest_ind(positive_credit_A8, negative_credit_A8)
statistic, pvalue
```
